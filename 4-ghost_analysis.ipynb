```python
import pickle
import matplotlib.pyplot as plt

plt.rcParams["font.size"] = 10 # 全体のフォントサイズが変更されます。
plt.rcParams['figure.figsize'] = (5.0, 3.5) 
plt.rcParams['figure.dpi'] = 300

# pickleファイルのパス
pickle_file_path = '/Prove/Yoshihisa/HEIF_ghost/PKL/pkl_second/1_1stQP35_2ndQP24_CTU64_PRESETmedium.pkl'

# pickleファイルをバイナリモードで開いてデータを読み込む
with open(pickle_file_path, 'rb') as file:
    loaded_data = pickle.load(file)

# 読み込んだデータからMAE結果を取得
ghost_results, ghost_results_shifted = loaded_data


# タプル内のリストを抽出
original_mae = ghost_results
shifted_mae = ghost_results_shifted



# QPの範囲
qp_range = list(range(0, 52))  # QPの値は1から始まると仮定しています

plt.figure

# MAE結果の折れ線グラフをプロット
plt.plot(qp_range, original_mae, label='Original')  # オリジナルのMAEをプロット
plt.plot(qp_range, shifted_mae, label='Circular shifted')  # シフトしたMAEをプロット
plt.xlabel('QP Value', fontsize=14)
plt.ylabel('Mean Absolute Error', fontsize=14)
# plt.title('Ghost Results')

plt.ylim(0, 18)
legend = plt.legend(loc='upper left')


# 凡例のフォントサイズを変更
for label in legend.get_texts():
    label.set_fontsize(10)  # フォントサイズを設定

plt.grid(True)
plt.xticks(range(0, 52, 5))
plt.savefig('double_MAE.png', bbox_inches="tight", pad_inches=0.0)
plt.show()

# MAEの差を計算して新しいリストに格納
mae_differences = [shifted - original for original, shifted in zip(original_mae, shifted_mae)]
plt.figure

# MAE差の折れ線グラフをプロット
plt.plot(qp_range, mae_differences, label='Difference')  # MAE差をプロット
plt.xlabel('QP Value', fontsize=14)
plt.ylabel('MAE Difference', fontsize=14)
# plt.title('MAE Differences between Original and Shifted')
legend = plt.legend(loc='upper left')

plt.ylim(-1.0, 2.5)
# plt.yticks(range(-1, 4))  # Y軸を小数点なしで設定

# 凡例のフォントサイズを変更
for label in legend.get_texts():
    label.set_fontsize(10)  # フォントサイズを設定
plt.grid(True)
plt.xticks(range(0, 52, 5))
plt.savefig('double_ghost.png', bbox_inches="tight", pad_inches=0.0)
plt.show()
```
